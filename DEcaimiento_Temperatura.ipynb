# Decaimiento de Temperatura

A continuación, tenemos los datos que fueron tomados con un termómetro en intervalos de 2 minutos, la temperatura del ambiente era de 22.1 ºC en ese momento.

<table><tr><td>Minutos </td><td>Temperatura</td></tr><tr><td>0 </td><td>79.2</td></tr><tr><td>2 </td><td>73.5</td></tr><tr><td>4 </td><td>69.5</td></tr><tr><td>6 </td><td>63.5</td></tr><tr><td>8 </td><td>60.8</td></tr><td>10 </td><td>58.5</td></table>

Los datos que usaremos para modelar nuestra ecuación serán los siguientes:
Temperatura del ambiente: 22.1 º C
Temperatura al tiempo 0: 79.2 º C
Temperatura al tiempo 4: 69.5 º C


Importamos nuestras librerias:

```python
from math import e, pow
import pandas as pd
import matplotlib.pyplot as plt
```

Establecemos nuestras listas:

```python
# Datos Experimentales.
d_exp = [79.2, 73.5, 69.5, 63.5, 60.8, 58.5]
t_exp = [0, 2, 4, 6, 8, 10]

# Datos teóricos.
temp_dec = []
avance_t = []
```

Generamos los datos teoricos:

```python
for tiempo in range(0, 180, 2):
    k_variable = float(-.0465)
    t_m = float(22.1)
    c = float(57.1)
    k_t = k_variable * tiempo
    temp_dec.append(round(float(t_m + c * pow(e, k_t)), 2))
    avance_t.append(float(tiempo))
```

Generamos la grafica:

```python
# Grafico.
plt.style.use("dark_background")
plt.title("Decaimiento de la temperatura, lata.")
plt.plot(t_exp, d_exp, label='Temperatura Experimental', color='m')
plt.plot(t_exp, d_exp, 'ro', label='Puntos de desfase')
plt.plot(avance_t, temp_dec, label='Temperatura teórica')
plt.ylabel('Temperatura (ºC)')
plt.xlabel("Tiempo (min)")
plt.legend()
plt.grid(True)
plt.show()
```

Opcional: Generamos un archivo CSV donde guardaremos nuestros datos teóricos.

```python
df = pd.DataFrame({'Tiempo': avance_t, 'Temperatura': temp_dec})
df.to_csv('Decaimiento_temperatura.csv', index=False) # Cambie a True para generar el archivo.
```
